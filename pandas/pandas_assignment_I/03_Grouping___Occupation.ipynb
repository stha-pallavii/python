### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

### Step 3. Assign it to a variable called users.

```python
users = pd.read_csv("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user", sep = "|")
```

```python
users.head()
```

### Step 4. Discover what is the mean age per occupation

```python
# users.groupby("occupation")["age"].mean()

users.groupby("occupation").age.mean()
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
#create new column 'is_male' that has value 'True' if gender='M'
users['is_male'] = users.gender.apply(lambda x: True if x == 'M' else False)
users[users['is_male']]
```

```python
# per occupation, divide no. of is_male by total gender count to get proportion of male
(users.groupby('occupation')["is_male"].sum() / users.groupby('occupation')["gender"].count()).sort_values(ascending = False)
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
# users.groupby("occupation")["age"].agg([min, max])

users.groupby("occupation").age.agg([min, max])
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
users.groupby(["occupation", "gender"])["age"].mean()
```

### Step 8.  For each occupation present the percentage of women and men

```python
# count total number of people in each occupation
occupation_count = users.groupby(['occupation']).count()

# count number of male and female in each occupation
occupation_by_gender = users.groupby(['occupation', 'gender']).agg({'gender': 'count'})

# divide the occupation_by_gender by the occupation_count and multiply by 100 to get percentage of each gender
occup_gender_percent = occupation_by_gender.div(occupation_count, level = "occupation")
occup_gender_percent.loc[:, 'gender']
```
