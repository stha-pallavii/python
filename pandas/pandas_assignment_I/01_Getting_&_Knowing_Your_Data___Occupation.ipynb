### Step 1. Import the necessary libraries

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

```python
#import csv   pd.read_csv("url", sep = "|")
dataset = pd.read_csv("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user", sep = "|")
```

### Step 3. Assign it to a variable called users and use the 'user_id' as index

```python
users = dataset.set_index("user_id")
```

### Step 11. Print only the occupation column

```python
#subsetting single column   dfname["column_name"]
print(users["occupation"])
```

### Step 12. How many different occupations are in this dataset?

```python
#first, create a subset by dropping duplicate values of occupation
unique_occupations = users.drop_duplicates(subset="occupation")

#count occupation column from the subset using  .value_counts() method
unique_occupations_count = unique_occupations["occupation"].value_counts()

#use len() method
print(len(unique_occupations_count))
```

### Step 13. What is the most frequent occupation?

```python
#first count the number of values of each occupation
occupation_count = users["occupation"].value_counts()

#sort the values in descending order and print the top value
print((occupation_count.sort_values(ascending=False)).head(1))
```
